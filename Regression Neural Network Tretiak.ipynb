# Regression Neural Network

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
df = pd.read_csv('marketing_prep.csv', sep=';')
```

```python
# Splitting the dataset into the Training set and Test set
X = df.iloc[:, :-7].values
y = df.iloc[:, 12:].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
from sklearn.preprocessing import StandardScaler
scalerX=StandardScaler().fit(X_train)
scalerY=StandardScaler().fit(y_train)
X_train=scalerX.transform(X_train)
y_train=scalerY.transform(y_train)
X_test=scalerX.transform(X_test)
y_test=scalerY.transform(y_test)
```

```python
# Install Tensorflow
# Install Keras
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input
```

```python
# Initialising the ANN 
'''
Побудуємо нейронну мережу з архітектурою 6-6-1
'''
rnn1 = Sequential()

# Adding the input layer and the first hidden layer
rnn1.add(Dense(output_dim = 6, activation = 'tanh', input_dim = 6))

# Adding the output layer
rnn1.add(Dense(1, activation='linear'))

# Compiling the ANN
rnn1.compile(optimizer='adam', loss='mean_squared_error', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
rnn1.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

```python
# Predict the Training set results
y_modeled1=rnn1.predict(X_train)
```

```python
# Predicting the Test set results
y_pred1 = rnn1.predict(X_test)
```

```python
# Initialising the ANN 
'''
Побудуємо нейронну мережу з архітектурою 6-6-6-1
'''
rnn2 = Sequential()

# Adding the input layer and the first hidden layer
rnn2.add(Dense(output_dim = 6, activation = 'tanh', input_dim = 6))

# Adding the hidden layer
rnn2.add(Dense(6, activation='tanh'))

# Adding the output layer
rnn2.add(Dense(1, activation='linear'))

# Compiling the ANN
rnn2.compile(optimizer='adam', loss='mean_squared_error', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
rnn2.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

```python
# Predict the Training set results
y_modeled2=rnn2.predict(X_train)
```

```python
# Predicting the Test set results
y_pred2 = rnn2.predict(X_test)
```

```python
# Initialising the ANN 
'''
Побудуємо нейронну мережу з архітектурою 6-6-6-6-1
'''
rnn3 = Sequential()

# Adding the input layer and the first hidden layer
rnn3.add(Dense(output_dim = 6, activation = 'tanh', input_dim = 6))

# Adding the hidden layer
rnn3.add(Dense(6, activation='tanh'))

# Adding the hidden layer
rnn3.add(Dense(6, activation='tanh'))

# Adding the output layer
rnn3.add(Dense(1, activation='linear'))

# Compiling the ANN
rnn3.compile(optimizer='adam', loss='mean_squared_error', metrics = ['accuracy'])
```

```python
# Predict the Training set results
y_modeled3=rnn3.predict(X_train)
```

```python
# Predicting the Test set results
y_pred3 = rnn3.predict(X_test)
```

```python
# Predict the Training set results
y_modeled4=rnn4.predict(X_train)
```

```python
# Predicting the Test set results
y_pred4 = rnn4.predict(X_test)
```

```python
# Initialising the ANN 
'''
Побудуємо нейронну мережу з архітектурою 6-6-4-1
'''
rnn5 = Sequential()

# Adding the input layer and the first hidden layer
rnn5.add(Dense(output_dim = 6, activation = 'tanh', input_dim = 6))

# Adding the hidden layer
rnn5.add(Dense(4, activation='tanh'))

# Adding the output layer
rnn5.add(Dense(1, activation='linear'))

# Compiling the ANN
rnn5.compile(optimizer='adam', loss='mean_squared_error', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
rnn5.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

```python
# Predict the Training set results
y_modeled5=rnn5.predict(X_train)
```

```python
# Predicting the Test set results
y_pred5 = rnn5.predict(X_test)
```

```python
# Inversing scaling
from sklearn.preprocessing import StandardScaler
y_modeled1 = scalerY.inverse_transform(y_modeled1)
y_modeled2 = scalerY.inverse_transform(y_modeled2)
y_modeled3 = scalerY.inverse_transform(y_modeled3)
y_modeled4 = scalerY.inverse_transform(y_modeled4)
y_modeled5 = scalerY.inverse_transform(y_modeled5)

y_pred1 = scalerY.inverse_transform(y_pred1)
y_pred2 = scalerY.inverse_transform(y_pred2)
y_pred3 = scalerY.inverse_transform(y_pred3)
y_pred4 = scalerY.inverse_transform(y_pred4)
y_pred5 = scalerY.inverse_transform(y_pred5)

y_train = scalerY.inverse_transform(y_train)
X_train = scalerX.inverse_transform(X_train)
y_test = scalerY.inverse_transform(y_test)
X_test = scalerX.inverse_transform(X_test)
```

```python
# Mean squared error
from sklearn.metrics import mean_squared_error
print("Mean Squared Error:")

print("6-6-1 RNN")
print(mean_squared_error(y_modeled1, y_train), mean_squared_error(y_pred1, y_test))

print("6-6-6-1 RNN")
print(mean_squared_error(y_modeled2, y_train), mean_squared_error(y_pred2, y_test))

print("6-6-6-6-1 RNN")
print(mean_squared_error(y_modeled3, y_train), mean_squared_error(y_pred3, y_test))

print("6-4-1 RNN")
print(mean_squared_error(y_modeled4, y_train), mean_squared_error(y_pred4, y_test))

print("6-6-4-1 RNN")
print(mean_squared_error(y_modeled5, y_train), mean_squared_error(y_pred5, y_test))
```

Найкращою з отриманих моделей нейронної мреежі є модель з архітекрутою 6-6-4-1. Адже в неї найменші значення середньоквадратичних помилок як на тренувальній, так і на тестовій вибірках.
Порівняємо дану модель з моделлю випадкового лісу. Для останньої похибки становили 0.4603978461663018 0.5306006805366204 для тренувальної та тестової вибірок відповідно. Як бачимо, середньоквадратична похибка моделі випадкового лісу для тренувальної вибірки є менше, ніж в моделі нейронної мережі, а от у випадку з тестовою, ситуація протилежна. Враховуючи, що модель нейронної мережі є менш перенавченою і що вона найкраще описує тестову вибірку, то можна зробити висновок про доцільність використання нейронної мережі з архітектурою 6-6-4-1 для апроксимації та передбачення трат користувачів магазину.
Зараз для нашої найкращої мережі відобразимо результати тренування та тестування
